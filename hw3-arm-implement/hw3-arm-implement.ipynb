```python
def APRIORI(data, min_support = 0.04,  max_length = 4):
    # Collecting Required Library
    import numpy as np
    import pandas as pd
    from itertools import combinations
    
    support = {} 
    L = list(data.columns)
    length = 1
    # write your code here !!
    can = combinations(L,1)
    #can 是指所有L裡面可能的column
    while (length <= max_length):#條件長度要小於max_length
        temp = {}
        for i in list(can):
            Sup = 0
            conditions = data.copy()
            for j in range(len(i)):
                conditions  = conditions[conditions[i[j]]==1]
                #這邊是指要取出符合條件的dataframe出來 例如說要計算同時購買bread跟coffee就要找出bread=1&coffee=1的dataframe
            Sup = len(conditions)/len(data) #計算support
            if Sup> 0.04:
                temp[i] = Sup #大於0.04就放入temp裡面
        support.update(temp)#把新加入的frequent itemset 跟對應的support放入 support 的dict裡面
        can = list()#新的候選人list
        obj = list(temp.keys())
        #當長度要往上加的時候就從原本長度裡面符合條件的開始找 因此obj 就是裝有前一個長度的frequent itemset
        for index1 in range(len(obj)):
            for index2 in range(index1, len(obj)):
                if  obj[index1]!= obj[index2] and len(set(obj[index1]) & set(obj[index2])) == length-1 and (tuple(set(obj[index1]) | set(obj[index2])) not in can):
                    #當index不一樣時並且兩者的只有一個item不一樣`,而且其聯集沒出現在can裡面時
                    selfjoin = tuple(set(obj[index1]) | set(obj[index2]))
                    #聯集
                    sub = combinations(list(selfjoin),length)
                    for i in list(sub):
                        if i not in list(temp.keys()):
                            break
                    #如果selfjoin裡面沒有含有infrequent itemset
                    can.append(selfjoin)
                    #才能加入候選人list裡面
        length+=1
        #update the length
    result = pd.DataFrame(list(support.items()), columns = ["Items", "Support"])
    return result
my_freq_itemset = APRIORI(transaction, 0.04, 3)
my_freq_itemset.sort_values(by = 'Support', ascending = False)
```

```python
def ASSOCIATION_RULE(df, min_confidence_threshold = 0.5):
    import pandas as pd
    from itertools import permutations
    
    support = pd.Series(df.Support.values, index=df.Items).to_dict()
    data = {"antecedents":[],"consequents":[],"consequent support":[],"antecedent support":[],"consequent support":[],
            "support":[],"confidence":[],"Lift":[],"Leverage":[],"Convection":[]}
    #先把需要的column都寫好
    L = df.Items.values
    #L裝有所有frequent itemsets 
    # write your code here !!
    for i in L:
        if len(i)>1:
            per = permutations(list(i),1)
            #長度大於1 而且必須要做排列 if a then b 跟 if b then a是不一樣的事情
            for j in per:
                other = tuple(set(i)-set(j))
                #other 就是consequents
                if support[i]/support[j] >= min_confidence_threshold:#當confidence大於min threshold時
                    confi = support[i]/support[j] #計算confidence
                    data["antecedents"].append(j) #update antecedents
                    data["consequents"].append(other)#update consequents
                    data["antecedent support"].append(support[j])#update antecedent
                    data["consequent support"].append(support[other])#update consequent support
                    data["support"].append(support[i])#update support
                    data["confidence"].append(confi)#update confidence
                    data["Lift"].append(confi/support[other])#update lift
                    data["Leverage"].append(support[i]-support[j]*support[other])#update leverage
                    data["Convection"].append((1-support[other])/(1-confi))#update convection
            
    result = pd.DataFrame(data, columns = ["antecedents", "consequents", "antecedent support", "consequent support",
                                        "support", "confidence", "Lift", "Leverage", "Convection"])
    return result
```

```python
## Rule with minimun confidence = 50%
my_rule = ASSOCIATION_RULE(my_freq_itemset, 0.5)
my_rule
# the column names in the following table is the same as those generated by mlxtend
# you can refer to http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/ 
# to find the formulas of lift, leverage, convection
```

```python
## finding frequent itemset with min support = 4%
frequent_itemset = apriori(df = transaction, min_support= 0.04, use_colnames= True)
frequent_itemset.sort_values(by = 'support', ascending = False)
```

```python
## Rule with minimun confidence = 50%
Rules = association_rules(frequent_itemset, min_threshold= 0.5)
Rules
```
